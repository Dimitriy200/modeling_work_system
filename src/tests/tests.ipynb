```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


df = pd.DataFrame(
    [
        [1, 1, 1, 0],
        [1, 2, 1, 0],
        [1, 2, 2, 0],
        [1, 1, 3, 0],
        [1, 5, 7, 1],
        [1, 6, 9, 1],
        
        [2, 10, 10, 0],
        [2, 11, 12, 0],
        [2, 11, 13, 0],
        [2, 15, 18, 1],
        [2, 17, 16, 1],
        
        [3, 2, 1, 0],
        [3, 2, 1, 0],
        [3, 2, 1, 0],
        [3, 2, 1, 1],
    ])

df.columns = ['group', 'feature1', 'feature2', 'anomaly']

df_normaly = df.query('anomaly == 0')
df_anomalous =  df.query('anomaly == 1')
```

```python
scaller = StandardScaler()
scaller_normaly = StandardScaler()
scaller_anomalous = StandardScaler()
```

```python
scaller.fit(df_normaly)
df_normaly_scaled = scaller.transform(df_normaly)
df_anomalous_scaled = scaller.transform(df_anomalous)
```

```python
df_normaly_scaled = pd.DataFrame(df_normaly_scaled)

df_normaly_scaled
```
